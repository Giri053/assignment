```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Reading the dataset to python

```python
data = pd.read_csv('iris.csv')
data.head()
```

# Columns in the dataset

```python
print("Columns in the dataset are:")
data.columns
```

# Mean of each column of the dataset

```python
mean_data=data.mean(axis=0,numeric_only=True)
print("The Mean value of each column in dataset are:")
mean_data
```

# Check for the null values present in the dataset

```python
data_null=data.isnull().sum()
print("The null values present in the dataset:")
data_null
```

# Visualizations using the iris dataset

# scatterplot

```python
sns.FacetGrid(data, hue="Classification", palette="hls", size=5) \
   .map(plt.scatter, "SL", "SW") \
   .add_legend()
plt.title("Comparison between various classification based on sapel length and width")
```

Data Insights:
Sepal width for setosa is high and length is low,Versicolor have average values for for sepal dimensions and
Virginica has small width but large sepal length.

# boxplot

```python
sns.boxplot(x="Classification", y="PL", palette="husl",data=data)
plt.title("Iris classification  based upon petal length",fontsize=16)
```

Data Insights:
Iris classification is clearly separable based upon petal length, with just a bit of potential overlap in the Iris-versicolor and Iris-virgnica.

# pairplot

```python
sns.pairplot(data, hue="Classification",palette="husl",  size=4, diag_kind="kde")
```

Data Insights:
    1.High co relation between petal length and width columns.
    2.Setosa has both low petal length and width
    3.Versicolor has both average petal length and width
    4.Virginica has both high petal length and width.
    5.Sepal width for setosa is high and length is low.
    6.Versicolor have average values for for sepal dimensions.
    7.Virginica has small width but large sepal length

# Checking Correlation

```python
plt.figure(figsize=(10,11))
sns.heatmap(data.corr(),annot=True)
plt.plot()
```

Data Insights:
Sepal Length and Sepal Width features are slightly correlated with each other

# Plotting the Histogram

```python
sns.FacetGrid(data, hue="Classification", height=5) \
.map(sns.distplot, "PW") \
.add_legend()
plt.title("Iris classification  based upon petal width",fontsize=16)
plt.show()
```

Data Insights:
petal width is a good Classification feature.The overlap is significantly less between Versicolor and Virginica,Setosa is well separated from the rest two.
